导包sqlite3


```python
import sqlite3 as sqlite3
```

读取table  
 pd.read_sql("sql语句", con)

```python
con = sqlite3.connect('./1703.sqlite')

df = pd.read_sql('select "Date/Time","Weather" from weather_2017',con)
df.head()
```

设置行索引index_col

```python
pd.read_sql('select * from weather_2017',con,index_col='Date/Time').shape
```
